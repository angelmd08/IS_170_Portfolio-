```python
import pandas as pd
df = pd.read_csv("healthcare_data-2.csv")
df
```

```python
df.dropna()
```

```python
from pandas.core.arrays.sparse import array
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
label = le.fit_transform(df['Outcome'])
label
df.drop('Outcome',axis = 1, inplace = True)
df['Outcome']= label
df
```

```python
from sklearn import preprocessing
df_scaled = preprocessing.scale(df)
df_scaled = pd.DataFrame(df_scaled,columns=df.columns)
df_scaled['Outcome']=df['Outcome']
df=df_scaled
df
```

```python
print(df.rename(columns={'BloodPressure':'BP'}))
print(df.rename(columns={'DiabetesPedigreeFunction':'Daibetes'}))
```

```python
from numpy import column_stack
old_cols=df.columns.values
new_cols= ['Pregnancies','BloodPressure','Glucose','BMI']
df=df.reindex(columns=new_cols)
df
```
