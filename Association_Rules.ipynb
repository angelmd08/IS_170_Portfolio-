```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#reading the data
df=pd.read_csv('GroceryStoreDataSet.csv')
df.head()
```

```python
# missing a header we'll create one called 'Products' 
df=pd.read_csv('GroceryStoreDataSet.csv', names=['products'])
df.head() 
```

```python
df.shape
```

```python
#converting dataframe into a list of lists.
data=list(df["products"].apply(lambda x:x.split(",")))
data
```

```python
#doing one hot encoding
#Each column represents an item and each row represent a record or a transaction. 
#items were given TRUE if it has occured and FALSE if it didnt 
#converted the str into a numeric value 

from mlxtend.preprocessing import TransactionEncoder
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)

#Turn False & True into numeric value
df=df.replace(False,0)
df=df.replace(True,1)

df
                
```

```python
#we did .1 support value threshold 
# The parameter (use_colnames=True) is used to convert the integer back to its respective name.

df=apriori(df,min_support=0.1,use_colnames=True)
df
```

With the SUPPORT values above, you can try and match your own manual calculation. You notice how some are grouped with two. ALL items have been tested alone and in groups, but only these ones with the SUPPORT threshold 0.2 or higher values (i.e. other groups didn't make 0.2 therefore not shown).

```python
#Let's view our interpretation values using the Association rule function.
df_ar=association_rules(df,metric="lift", min_threshold=1)
df_ar
```

From above result, we say that the support is 0.2 calculated by dividing the number of transactions containing milk and bread by the total number of transactions.  The confidence level is 0.30 shows that out of all the transactions that contain bread 0.30 contain milk too.  The lift 1.23 tells us that milk is 1.23 times more likely to be bought by the customer who also buys bread.

```python
#Let's view our interpretation values using the Association rule function.
df_ar=association_rules(df,metric="confidence", min_threshold=0.6)
df_ar
```

From above result, let's take a look at the first row items: milk and bread. 
The two items have the support of 0.20, where milk is 0.25 and bread is 0.65. 
confidence is 0.8, this implies that if a person bought milk then there is 80% bread is also bought. There are also positive correlations as we can see from the positive values of lift, leverage, and conviction.
